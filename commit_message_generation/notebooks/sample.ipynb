```python
import os
import pandas as pd
from huggingface_hub import list_repo_tree, hf_hub_download


models = [os.path.basename(_.path) for _ in list_repo_tree("JetBrains-Research/lca-results", repo_type="dataset", path_in_repo="commit_message_generation/predictions")]
models_predictions = {}
for model in models:
    try:
        local_path = hf_hub_download("JetBrains-Research/lca-results", f"commit_message_generation/predictions/{model}/predictions.jsonl", repo_type="dataset")
    except Exception as e:
        prediction_file = [_.path for _ in list_repo_tree("JetBrains-Research/lca-results", repo_type="dataset", path_in_repo=f"commit_message_generation/predictions/{model}")][0]
        local_path = hf_hub_download("JetBrains-Research/lca-results", prediction_file, repo_type="dataset")
    df = pd.read_json(local_path, orient="records", lines=True)
    df = df.set_index(["repo", "hash"])
    models_predictions[model] = df

local_path = hf_hub_download("JetBrains-Research/lca-commit-message-generation", "commitchronicle-py-long/test-00000-of-00001-c91b0bca349ab4a3.parquet", repo_type="dataset")
df = pd.read_parquet(local_path)
df = df.set_index(["repo", "hash"])
```

```python
import random

random.seed(42)

NUM_SAMPLES = 500

sampled_data = []
already_sampled = set()
for _ in range(NUM_SAMPLES):
    sample_pair = random.sample(models, 2)
    sample_commit = random.sample(df.index.tolist(), 1)[0]
    pred_A = models_predictions[sample_pair[0]].loc[sample_commit]["prediction"]
    pred_B = models_predictions[sample_pair[1]].loc[sample_commit]["prediction"]
    commit = df.loc[sample_commit]

    if (sample_pair[0], sample_pair[1], sample_commit[0], sample_commit[1]) in already_sampled:
        continue
    
    sampled_data.append({"model_A_pred": pred_A, 
                         "model_B_pred": pred_B, 
                         "repo": sample_commit[0], 
                         "hash": sample_commit[1], 
                         "url": f"https://github.com/{sample_commit[0]}/commit/{sample_commit[1]}",
                         "model_A": sample_pair[0], 
                         "model_B": sample_pair[1]})
    
    already_sampled.add((sample_pair[0], sample_pair[1], sample_commit[0], sample_commit[1]))

sampled_df = pd.DataFrame(sampled_data)
sampled_df["id"] = sampled_df.index
sampled_df.head()
```

```python
from huggingface_hub import upload_file, create_repo
from dotenv import load_dotenv

load_dotenv()

sampled_df.to_json("sampled_data.jsonl", orient="records", lines=True)
try:
    upload_file(
        repo_id="saridormi/lca-cmg-labels",
        repo_type="dataset",
        path_or_fileobj="sampled_data.jsonl",
        path_in_repo="sampled_data.jsonl",
    )
except Exception as e:
    create_repo(repo_id="saridormi/lca-cmg-labels", repo_type="dataset", private=True)
    upload_file(
        repo_id="saridormi/lca-cmg-labels",
        repo_type="dataset",
        path_or_fileobj="sampled_data.jsonl",
        path_in_repo="sampled_data.jsonl",
    )
```
